**Problem Statement 1:Smart Parking Lot Management System**

Design a function to manage a smart parking lot.
    
The system should:
Accept vehicle entry and exit logs

Calculate total parked vehicles

Identify peak parking usage

Alert if parking exceeds capacity


```python
def smart_parking(capacity, logs):
    parked = 0
    peak = 0
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1
        if parked > peak:
            peak = parked

    if parked > capacity:
        status = "Alert: Parking Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)
    print("Parking Status:", status)


capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

**Problem Statement 2:Online Food Delivery Time Estimator**

Create a function that estimates delivery time based on:
Distance (km)

Weather condition

Traffic level

Apply delays dynamically and display final ETA.



```python
def delivery_time(distance, traffic, weather):
    time = distance * 5

    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5

    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 15

    print("Estimated Delivery Time:", time, "minutes")


delivery_time(8, "High", "Rainy")
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

Build a function that analyzes seat booking data and:
Calculates occupancy percentage

Determines if show is Housefull

Suggests opening additional shows

```python
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)


total_seats = 200
booked_seats = [1]*150


seat_occupancy(total_seats, booked_seats)
```

**Problem Statement 4:Cloud Server Load Classification System**

Create a function to classify server load based on CPU usage readings.
Rules:
Average CPU < 50% → Normal

50%–80% → Warning

80% → Critical

```python
def server_load(cpu_readings):
    total = 0
    
    for value in cpu_readings:
        total += value
        
    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)

server_load([45, 60, 70, 85, 90])
```

**Problem Statement 5:Smart Classroom Resource Usage Monitor**

Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.
Real-Time Use
Smart classrooms

Energy optimization systems

```python
def resource_monitor(resources):
    overused = []

    for resource in resources:
        if resources[resource] > 8:
            overused.append(resource)

    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}


resource_monitor(usage)
```
