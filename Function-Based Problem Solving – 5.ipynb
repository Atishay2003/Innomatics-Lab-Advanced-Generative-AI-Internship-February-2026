**Problem 1: Social Media – Post Engagement Analyzer**

Create a function to analyze engagement on a social media post.


```python
def analyze_post(likes_list): 
    total_likes = 0
    for like in likes_list:
        total_likes = total_likes + like
    print("Total Likes:", total_likes)
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes = [200, 300, 150, 400]
analyze_post(likes)
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

Create a function that analyzes seasonal rainfall data.


```python
def check_rainfall(rainfall_data, required_level):   
    total = 0
    
    for rain in rainfall_data:
        total = total + rain
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", int(average))
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

season_data = [70, 75, 68, 80, 67]
required_level = 70

check_rainfall(season_data, required_level)
```

**Problem 4: Social Media – Duplicate Account Detection**

Create a function to detect duplicate usernames in a system.



```python
def detect_duplicates(usernames):
    unique_usernames = set(usernames) 
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users = ["Adj123", "asd45", "dsfc123", "dfc77"]
detect_duplicates(users)
```

**Problem 6: Agriculture – Premium Crop Price Filter**

Create a function that filters crop prices above a premium threshold.


```python
def filter_premium_crops(prices):
    
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)
crop_prices = [1500, 2500, 1800, 3200, 900]
filter_premium_crops(crop_prices)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

Create a function to calculate attendance percentage and check exam eligibility.


```python
def check_attendance(attendance_list):
    
    total_classes = len(attendance_list)
    present_count = 0
    
    for day in attendance_list:
        if day == 1:
            present_count += 1
    
    percentage = (present_count / total_classes) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = [1, 1, 1, 0, 1]
check_attendance(attendance)
```

**Problem 10: Smart Electricity Bill Analyzer**

You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.


```python
def calculate_electricity_bill(units):
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    
    return bill, usage


bill_amount, usage_status = calculate_electricity_bill(250)

print("Total Bill:", bill_amount)
print("Usage Level:", usage_status)
```
