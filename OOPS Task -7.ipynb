**Problem 1: Smart Light Controller**

Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class LightController:

    def __init__(self, light_name):
        self.light_name = light_name
        self.current_state = False

    def switch(self, action):
        if action.upper() == "ON":
            self.current_state = True
        elif action.upper() == "OFF":
            self.current_state = False

    def show(self):
        state = "ON" if self.current_state else "OFF"
        print(f"{self.light_name} is {state}")

name = "Bedroom Light"
action = "ON"

lamp = LightController(name)
lamp.switch(action)
lamp.show()

```

**Problem 2: Employee ID Card System**

Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.


```python
class StaffMember:

    def __init__(self, emp_name, emp_code, emp_dept):
        self.emp_name = emp_name
        self.emp_code = emp_code
        self.emp_dept = emp_dept

    def print_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.emp_name)
        print("ID:", self.emp_code)
        print("Department:", self.emp_dept)
name_input = input("Employee Name: ")
id_input = input("Employee ID: ")
dept_input = input("Department: ")

employee1 = StaffMember(name_input, id_input, dept_input)

employee1.print_card()
```

**Problem 3: Mobile Contact Record**

Problem Statement
Create a class to store contact name and phone number and display contact information.


```python
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")
c1 = Contact(contact_name, phone_number)
c1.show_contact()
```

**Problem 4: Product Price Tag Generator**

Problem Statement
Create a class that stores product name and price and prints a formatted price tag.


```python
class Product:

    def __init__(self, pname, price):
        self.pname = pname
        self.price = price

    def display_tag(self):
        print("Product:", self.pname)
        print("Price: ₹" + str(self.price))

product_name = input("Product Name: ")
product_price = input("Price: ")
item1 = Product(product_name, product_price)
item1.display_tag()
```

**Problem 5: Movie Rating Display System**

Problem Statement
Create a class that stores movie name and rating and displays movie details.


```python
class MovieInfo:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating
    def show_details(self):
        print("Movie:", self.title)
        print("Rating:", self.rating, "/ 5")
movie_name = input("Movie Name: ")
movie_rating = input("Rating: ")
m1 = MovieInfo(movie_name, movie_rating)
m1.show_details()
```

**Problem 6: Delivery Address Manager**

Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def show_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)
cust_name = input("Customer Name: ")
cust_address = input("Address: ")
d1 = Delivery(cust_name, cust_address)
d1.show_delivery()
```
