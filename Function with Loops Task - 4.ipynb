**Problem 1: Mobile Recharge Validation System**

Create a function to validate a mobile recharge request.


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be ₹50 or more.")
        return False
    
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False
    
    else:
        print("Recharge Successful!")
        return True

while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        
        if validate_recharge(amount):
            break   
        else:
            print("Please try again.\n")
            
    except ValueError:
        print("Please enter a valid number.\n")
```

**Problem 2: Inventory Reorder Alert System**

Given a dictionary of products and stock quantities:


```python
def check_inventory(products):
    for item, quantity in products.items():
        
        if quantity < 15:
            print(f"{item}: Reorder Alert")
            
        else:
            print(f"{item}: Stock OK")

current_inventory = {
    "laptop": 12, 
    "Mobile Phones": 45, 
    "Monitor": 8,
    "Keyboard": 14,
    "Mouse": 54
}

check_inventory(current_inventory)
```

**Problem 3: Student Result Processing System**

Create a function that:
Accepts a list of student marks

Calculates average marks

Prints Pass if average ≥ 50 else Fail


```python

def calculate_result(marks):
    total = 0
    for mark in marks:
        total = total + mark
    average = total / len(marks)
    
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    return average, result

students = {
    "Salman": [60, 70, 80],
    "Mishra": [40, 45, 50],
    "Sharma": [55, 85, 85],
    "Riya": [30, 35, 40]
}
for name, marks in students.items():
    
    avg, final_result = calculate_result(marks)
    
    print("Student Name:", name)
    print("Marks:", marks)
    print("Average:", avg)
    print("Result:", final_result)
    print("\n")
```

**Problem 4: Cab Fare Estimator with Retry Option**

Create a cab fare calculator:
Base fare = ₹50

₹12 per km

Peak hour → 25% extra

Allow retry using while loop


```python

def calculate_fare(distance, is_peak):
    
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if is_peak == "yes":
        fare = fare * 1.25   # 25% extra
    
    return fare

while True:
    
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (Yes/No): ").lower()
        
        total_fare = calculate_fare(distance, peak)
        
        print("Total Cab Fare: ₹", total_fare)
        
    except ValueError:
        print("Please enter valid distance.")
    
    # Retry option
    retry = input("Do you want to calculate again? (Yes/No): ").lower()
    
    if retry != "yes":
        print("Thanks Fare Used car Estimator")
        break
```

**Problem 5: Employee Attendance Eligibility Checker**

Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


```python
def check_attendance(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present = present + 1
    total_days = len(attendance)
    percentage = (present / total_days) * 100
     
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance_list = ["P", "A", "P", "P", "P", "A", "P"]
result = check_attendance(attendance_list)

print("Result:", result)
```

**Problem 6: Password Strength Checker**

Create a function to check password strength.
Rules:
Minimum length = 8

Must contain at least one digit

Must contain at least one special character (@#$)

Real-Time Use

```python
def check_password(password):

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"

user_password = input("Enter your password: ")

result = check_password(user_password)

print(result)
```
